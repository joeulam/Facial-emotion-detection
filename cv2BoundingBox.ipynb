```python
import cv2
import tensorflow as tf
import numpy as np

# Load the TensorFlow model
model_path = 'C:/Users/joeul/OneDrive/Documents/GitHub/Facial-emotion-detection/Emotion/emotion'
new_model = tf.keras.models.load_model(model_path)

# Open a video capture object
vid = cv2.VideoCapture(0)  # Use 0 for the default camera, adjust if using a different camera

# Emotion labels
emotion_labels = ['angry', 'disgusted', 'fearful', 'happy', 'neutral', 'sad', 'surprised']

# Load the face detection cascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
confidence_threshold = 2.5
while True:
    # Capture the video frame by frame
    ret, frame = vid.read()

    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces using the Haar cascade
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        # Extract the face region
        face_roi = frame[y:y + h, x:x + w]

        # Preprocess the face for the emotion model
        input_image = cv2.resize(face_roi, (180, 180))
        input_image = np.expand_dims(input_image, axis=0)  # Add batch dimension if needed

        # Run emotion prediction using your TensorFlow model
        predictions = new_model.predict(input_image)

        # Get the predicted emotion label and confidence score
        predicted_label = emotion_labels[np.argmax(predictions, axis=1).astype(int)[0]]
        confidence_score = predictions[0, np.argmax(predictions, axis=1)[0]]
        if confidence_score >= confidence_threshold:
            # Draw bounding box around the face
            cv2.rectangle(frame, (x, y), (x + w, y + h), (225, 0, 0), 2)

            # Display emotion label and confidence score
            label = f"Emotion: {predicted_label}, Confidence: {confidence_score:.2f}"
            cv2.putText(frame, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 225), 2)

    # Display the resulting frame
    cv2.imshow('frame', frame)

    # Check for the 'q' key to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close all windows
vid.release()
cv2.destroyAllWindows()
```
